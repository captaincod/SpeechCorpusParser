```python
from transformers import BertForMaskedLM,BertTokenizer, pipeline
model=BertForMaskedLM.from_pretrained('sberbank-ai/ruBert-base')
tokenizer=BertTokenizer.from_pretrained('sberbank-ai/ruBert-base', do_lower_case=False)
unmasker = pipeline('fill-mask', model=model,tokenizer=tokenizer)
```

# Предложения

Весёлые истории из жизни

```python
with open("fun_masked.txt") as file:
  sentences = file.read().splitlines()
  sentences = list(filter(None, sentences))
```

```python
with open("fun_masked.txt") as file:
  masked = file.read().splitlines()
  masked = list(filter(None, masked))
```

Рассказы сибиряков о жизни

```python
with open("sib_sentences.txt") as file:
  sentences = file.read().splitlines()
  sentences = list(filter(None, sentences))
```

```python
with open("sib_masked.txt") as file:
  masked = file.read().splitlines()
  masked = list(filter(None, masked))
```

# Предсказывание

```python
for i in range(len(masked)):
  try:
    output = f"{sentences[i]}\n"
    output += f"{masked[i]}\n"
    for i in unmasker(masked[i]):
      output += f"{i['sequence']}\n"
  except (TypeError):
    output = ""
  print(output, end="\n")
```
